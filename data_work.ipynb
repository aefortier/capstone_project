```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from variable_dict import county_codes
```

```python
census = pd.read_pickle('census_data_all.pkl')
election = pd.read_pickle('election_data_all.pkl')
```

```python
census['county'] = census['county'].replace(county_codes)
```

```python
new_census = census.loc[census['year'] != 2020]
```

```python
new_census['year'] = new_census['year'].astype(str)
election['year'] = election['year'].astype(str)
```

```python
new_census.loc[new_census['year'] == '2011', 'year'] = '2010'
```

```python
new_census = new_census.loc[new_census['year'].isin(['2010', '2012', '2014', '2016', '2018'])]
```

```python
election = election.merge(new_census, how = 'left', on = ['county', 'year', 'state'])
```

```python
election['winner'] = election['winner'].replace({
    'D' : 0,
    'R' : 1
})
```

```python
header_list = list(new_census)
header_list.remove('year')
header_list.remove('state')
header_list.remove('county')
```

```python
X_train = election[header_list]
y_train = election[['Democrat', 'Republican', 'Other']]
```

```python
election
```

```python
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train, y_train)
```

```python
data_2020 = census.loc[census['year'] == 2020]
```

```python
data_2020 = data_2020.reset_index(drop = True)
```

```python
X_pred = data_2020[header_list]
```

```python
predicted_winner = pd.DataFrame(clf.predict(X_pred))
```

```python
winner_2020 = predicted_winner.join(data_2020[['county', 'state', 'year']])
```

```python
winner_2020 = winner_2020.rename(columns = {0 : 'Democrat',
                                            1 : 'Republican',
                                            2 : 'Other'})
```

```python
election = election[['county', 'state', 'year', 'Other', 'Republican', 'Democrat', 'office']]
```

```python
full_election = pd.concat([election, winner_2020])
```

```python
full_election.to_excel('full_election.xlsx')
```
